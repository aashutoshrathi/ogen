```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Set up some basic parameters
DEPOSIT_AMOUNT = 10000              # POLIS
REWARD_PER_MINUTE = 5.2             # POLIS/min
BLOCK_DURATION = 30                 # seconds
EPOCH_LENGTH = 16                   # blocks
INACTIVITY_PENALTY_QUOTIENT = 1024  #
CURRENT_NUMBER_POLIS = 9831891      # POLIS
BASE_REWARD_QUOTIENT = 1024         #

BASE_REWARD_PER_BLOCK = REWARD_PER_MINUTE * 60 / BLOCK_DURATION # POLIS/block
BASE_REWARD_PER_EPOCH = BASE_REWARD_PER_BLOCK * EPOCH_LENGTH    # POLIS/epoch

print(f"{BASE_REWARD_PER_BLOCK} POLIS/block")
print(f"{BASE_REWARD_PER_BLOCK} POLIS/epoch")
print(f"{REWARD_PER_MINUTE} POLIS/min")
print(f"{REWARD_PER_MINUTE * 60 * 24} POLIS/day")
print(f"{REWARD_PER_MINUTE * 60 * 24 * 365} POLIS/year")
```

```python
validators = np.arange(10, 1000, 10)
reward = BASE_REWARD_PER_EPOCH / validators

plt.ylabel("Inflation (POLIS)")
plt.xlabel("Number of Validators")
plt.title("Validator Rewards based on Number of Validators")
plt.plot(validators, reward)

None
```

```python
polis = CURRENT_NUMBER_POLIS
current_reward = REWARD_PER_MINUTE
supply = []
for i in range(24 * 365 * 20):
    supply.append(polis)
    polis += current_reward * 60
    if i % 730 == 0:
        polis += 55296
    if i % (24 * 365) == 0:
        current_reward /= 5

supply = np.array(supply)
inflation = REWARD_PER_MINUTE * 60 * 24 * 365 / supply * 100

plt.xlabel("Years in the future")
plt.ylabel("Inflation rate (yearly)")
plt.title("Annualized inflation over time")
x = np.arange(0, 20, 1/(24*365))
plt.plot(x, inflation)
None
```

```python
def calculate_penalties_over_time(percent_voting = 0.1):
    voting_validator_balance = DEPOSIT_AMOUNT
    nonvoting_validator_balance = DEPOSIT_AMOUNT
    
    voting_balances = []
    nonvoting_balances = []
    ratio = []
    
    for i in range(4, 200):
        # every epoch:
        base_reward = BASE_REWARD_PER_BLOCK
        nonvoting_validator_balance -= base_reward
        nonvoting_validator_balance -= nonvoting_validator_balance * i / INACTIVITY_PENALTY_QUOTIENT
        
        voting_balances.append(voting_validator_balance)
        nonvoting_balances.append(nonvoting_validator_balance)
        
        total_voting_balance = voting_validator_balance * percent_voting
        total_nonvoting_balance = nonvoting_validator_balance * (1 - percent_voting)
        ratio.append(total_voting_balance / (total_nonvoting_balance + total_voting_balance))
    
    return np.array(voting_balances), np.array(nonvoting_balances), np.array(ratio)

voting_bals, nonvoting_bals, ratios = calculate_penalties_over_time()

plt.title("Inactivity Leak for no finality for Voting and Non-Voting Validators")
plt.xlabel("Epochs since Finality")
plt.ylabel("Balance (POLIS)")
plt.plot(voting_bals)
plt.plot(nonvoting_bals)
plt.show()

plt.title("Inactivity Leak during no finality ratio of voting to non-voting")
plt.xlabel("Epochs since Finality")
plt.ylabel("Voting / Non-Voting")
plt.plot(ratios)
plt.show()

None
```

```python
def calculate_penalties_over_time_single(validators = 500):
    voting_validator_balance = 0
    nonvoting_validator_balance = 0
    
    voting_balances = []
    nonvoting_balances = []
    
    for i in range(0, 2000):
        # every epoch:
        base_reward = BASE_REWARD_PER_BLOCK / validators
        voting_validator_balance += base_reward
        nonvoting_validator_balance -= base_reward
        
        voting_balances.append(voting_validator_balance)
        nonvoting_balances.append(nonvoting_validator_balance)
        
        total_voting_balance = voting_validator_balance
        total_nonvoting_balance = nonvoting_validator_balance
    
    return np.array(voting_balances), np.array(nonvoting_balances)

num_validators = np.arange(500, 2000, 100)

voting_bal, nonvoting_bal = calculate_penalties_over_time_single(num_validators)

voting_bal.shape

plt.xlabel("Num epochs")
plt.ylabel("Balance")
plt.imshow(voting_bal, x)
```
